```python
import intake
import pandas as pd
import re
```

```python
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

```python
df_cordex = cat['c3s-cordex'].read()
df_cordex_original = df_cordex.copy()
df_cordex
```

## [2] add missing file

eur11-hist-day-mohc_hadgem2_es-knmi_racmo22e-r1i1p1

Missing rlds variable (surface downwelling longwave flux) data for 1981-1985.

```python
df_2 = df_cordex.loc[
    (df_cordex.variable=="rlds") 
    & (df_cordex.experiment_id=="historical")
    & (df_cordex.rcm_name=="KNMI-RACMO22E")
    & (df_cordex.domain=="EUR-11")
    & (df_cordex.time_frequency=="day")
    & (df_cordex.driving_model=="MOHC-HadGEM2-ES")
    & (df_cordex.ensemble=="r1i1p1")
]
df_2.head()
```

```python
# Copy first matching row
new_row = df_2.iloc[0].copy()

# Step: set size

new_row["size"] = 764239426

# Step: Set new times
new_row["start_time"] = "1981-01-01T12:00:00"
new_row["end_time"]   = "1985-12-30T12:00:00"

# Step: Update path
old_path = new_row["path"]

# Build new date string
new_date_str = "19810101-19851230"

# Replace old date in filename (assumes format: ..._<dates>.nc)
new_path = re.sub(r"\d{8}-\d{8}", new_date_str, old_path)

# Assign updated path
new_row["path"] = new_path

# Step: Append to df_cordex
df_cordex = pd.concat([df_cordex, pd.DataFrame([new_row])], ignore_index=True)

# check new row
df_cordex.tail(1)
```

```python
# Copy first matching row
new_row = df_3.iloc[0].copy()

# Step: set size

new_row["size"] = 520680447

# Step: Set new times
new_row["start_time"] = "2071-01-01T12:00:00"
new_row["end_time"]   = "2075-12-30T12:00:00"

# Step: Update path
old_path = new_row["path"]

# Build new date string
new_date_str = "20710101-20751230"

# Replace old date in filename (assumes format: ..._<dates>.nc)
new_path = re.sub(r"\d{8}-\d{8}", new_date_str, old_path)

# Assign updated path
new_row["path"] = new_path

# Step: Append to df_cordex
df_cordex = pd.concat([df_cordex, pd.DataFrame([new_row])], ignore_index=True)

# check new row
df_cordex.tail(1)
```

## [7] remove obsolte version v20190919

```python
df_7 = df_cordex.loc[
    (df_cordex.variable=="mrro") 
    & (df_cordex.experiment_id=="historical")
    & (df_cordex.institute=="CNRM")
    & (df_cordex.domain=="EUR-11")
    & (df_cordex.time_frequency=="6hr")
    & (df_cordex.rcm_name=="CNRM-ALADIN63")
    & (df_cordex.driving_model=="MOHC-HadGEM2-ES")
    & (df_cordex.ensemble=="r1i1p1")
    & (df_cordex.version=="v20190919")
]
df_7.head()
```

```python
# Define the filter condition for rows you want to remove
mask_to_remove_v20190919 = (
    (df_cordex.variable == "mrro") 
    & (df_cordex.experiment_id == "historical")
    & (df_cordex.institute == "CNRM")
    & (df_cordex.domain == "EUR-11")
    & (df_cordex.time_frequency == "6hr")
    & (df_cordex.rcm_name == "CNRM-ALADIN63")
    & (df_cordex.driving_model == "MOHC-HadGEM2-ES")
    & (df_cordex.ensemble == "r1i1p1")
    & (df_cordex.version == "v20190919")
)

# Remove the matching rows
df_cordex = df_cordex[~mask_to_remove_v20190919]
df_cordex
```

## [7] remove broken year 1950

```python
mask_to_remove_year_1950 = (
    (df_cordex.variable == "mrro") 
    & (df_cordex.experiment_id == "historical")
    & (df_cordex.institute == "CNRM")
    & (df_cordex.domain == "EUR-11")
    & (df_cordex.time_frequency == "6hr")
    & (df_cordex.rcm_name == "CNRM-ALADIN63")
    & (df_cordex.driving_model == "MOHC-HadGEM2-ES")
    & (df_cordex.ensemble == "r1i1p1")
    & (df_cordex.start_time == "1950-01-01T03:00:00")
)

# Remove the matching rows
df_cordex = df_cordex[~mask_to_remove_year_1950]
df_cordex
```

## Write new catalog

```python
from datetime import datetime, UTC

last_updated = datetime.now(UTC)
version = last_updated.strftime('v%Y%m%d')
cat_name = f"c3s-cordex_{version}.csv.gz"
cat_path = f"../intake/catalogs/c3s-cordex/{cat_name}"

df_cordex.to_csv(cat_path, index=False, compression="gzip")
```
