# Update cordex manifest file from intake catalog

```python
import intake
import pandas as pd
import datetime
```

```python
# cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/refs/heads/update-cordex-2025-07-11/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
cats = list(cat)
cats
```

## Update cordex manifest

```python
df_cordex = cat["c3s-cordex"].read()
```

```python
df_cordex
```

```python
cds_domain = "data.mips.climate.copernicus.eu"
base_url = f"https://{cds_domain}/thredds/fileServer/esg_c3s-cordex"

df_cordex["url"] = df_cordex["path"].apply(lambda x: f"{base_url}/{x}")
```

```python
list(df_cordex["url"])[0]
```

```python
last_updated = datetime.datetime.now(datetime.UTC)
version = last_updated.strftime('%Y%m%d')
manifest_path = f"../cordex/manifest_C3S_CORDEX_{version}.txt.gz"

df_cordex.to_csv(manifest_path, columns=["url"], header=False, index=False, compression="gzip")
```
