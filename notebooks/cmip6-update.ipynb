# CMIP6 Catalog

```python
import intake
import pandas as pd
```

```python
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"
#cat_url = "https://github.com/cp4cds/c3s_34g_manifests/raw/add-fixes-for-cmip6-cordex/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

## Load CMIP6 catalog

```python
df_cmip6 = cat['c3s-cmip6'].read()
df_cmip6
```

```python
df_cmip6.nunique()
```

## Load old CMIP6 release 3 catalog

```python
df_cmip6_r3 = cat['c3s-cmip6-r3'].read()
df_cmip6_r3
```

```python
df_cmip6_r3.nunique()
```

## Compare dataset IDs

```python
# added datasets in new version

set(df_cmip6.ds_id) - set(df_cmip6_r3.ds_id)
```
