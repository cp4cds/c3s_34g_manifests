# CMIP6 catalog: available grids


```python
import intake
import pandas as pd
```

```python
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

## Load cmip6 catalog

```python
df_cmip6 = cat['c3s-cmip6'].read()
df_cmip6
```

```python
df_cmip6.nunique()
```

```python
list(df_cmip6.grid_label.unique())
```

## group by grids

```python
ds_grids = {}

attrs = ["institution_id", "source_id", "experiment_id", "member_id"]

grouped = df_cmip6.groupby("grid_label")
for name, group in grouped:
    print(f"Grid: {name}")
    print("=========")
    print(group.nunique())
    print("\n")

    ds_grids[name] = {}
    
    for attr in attrs:
        values = list(group[attr].unique())
        values = [str(val) for val in values]
        values.sort()
        print(f"{attr}:\n {values}\n")

        ds_grids[name][attr] = set(values) 
    print("\n")
```
