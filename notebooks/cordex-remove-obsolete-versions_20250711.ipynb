```python
import intake
import pandas as pd
import re
```

```python
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

```python
df_cordex = cat['c3s-cordex'].read()
df_cordex_original = df_cordex.copy()
df_cordex
```

## Load list with obsolete versions

```python
df_excluded = pd.read_csv('../cordex/IPSL_C3S_CORDEX_obsolete_versions_20250709.txt.gz', names=["url"]).astype(str)
df_excluded
```

```python
print(df_excluded.url[0])
```

```python
prefix = "https://data.mips.climate.copernicus.eu/thredds/fileServer/esg_c3s-cordex/"

df_excluded["path"] = df_excluded["url"].str.replace(prefix, "", regex=False)
df_excluded.head()
```

```python
excluded_paths = set(df_excluded["path"])
len(excluded_paths)
```

## Remove obsolete versions from the catalog

```python
df_cordex = df_cordex[~df_cordex["path"].astype(str).isin(excluded_paths)]
df_cordex
```

## Write new catalog

```python
from datetime import datetime, UTC

last_updated = datetime.now(UTC)
version = last_updated.strftime('v%Y%m%d')
cat_name = f"c3s-cordex_{version}.csv.gz"
cat_path = f"../intake/catalogs/c3s-cordex/{cat_name}"

df_cordex.to_csv(cat_path, index=False, compression="gzip")
```

## compare catalog

```python
# Combine to find rows that differ between the two
diff = pd.concat([df_cordex, df_cordex_original]).drop_duplicates(keep=False)

# Rows in the new df that were not in the original (added)
added_rows = pd.concat([df_cordex, df_cordex_original, df_cordex_original]).drop_duplicates(keep=False)

# Rows in the original df that are no longer in the new one (removed)
removed_rows = pd.concat([df_cordex_original, df_cordex, df_cordex]).drop_duplicates(keep=False)
```
