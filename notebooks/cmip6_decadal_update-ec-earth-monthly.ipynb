# CMIP6 Decadal catalog: update EC-EARTH

Filter out the daily minimum/maximum near surface temperature from the monthly datasets, please? This is available only for the EC-EARTH model.

```python
import intake
import pandas as pd
```

```python
#cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/update-cmip6-decadal-ec-earth-2023-11-01/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

## Load cmip6 decadal catalog

```python
df_cmip6_decadal = cat['c3s-cmip6-decadal'].read()
df_cmip6_decadal
```

```python
df_cmip6_decadal.nunique()
```

```python
df_cmip6_decadal.institution_id.unique()
```

```python
df = df_cmip6_decadal.loc[df_cmip6_decadal.institution_id == 'EC-Earth-Consortium']
df.nunique()
```

```python
drop_condition = ((df_cmip6_decadal.institution_id=="EC-Earth-Consortium") &
    (df_cmip6_decadal.table_id=="Amon") &
    ((df_cmip6_decadal.variable_id=="tasmax") | (df_cmip6_decadal.variable_id=="tasmin"))
)


df_drop = df_cmip6_decadal.loc[drop_condition]
df_drop
```

```python
df_drop.nunique()
```

```python
df_new = df_cmip6_decadal.loc[~drop_condition]
df_new
```

```python
df = df_new.loc[df_new.institution_id == 'EC-Earth-Consortium']
df.nunique()
```

```python
df_new.nunique()
```

## Check variable per institution

```python
institutions_ids = list(set(df_new.institution_id))
institutions_ids
```

```python
for institution_id in institutions_ids:
    print(f"\n{institution_id}:")
    df_inst = df_new.loc[
        (df_new.institution_id==institution_id)
    ]
    table_ids = list(set(df_inst.table_id))
    table_ids.sort()
    for table_id in table_ids:
        df_month = df_inst.loc[(df_inst.table_id==table_id)]
        variables = list(set(df_month.variable_id))
        variables.sort()
        print(f"- {table_id}: {variables}")
```

## Write new catalog

```python
import datetime

last_updated = datetime.datetime.now(datetime.UTC)
version = last_updated.strftime('v%Y%m%d')
cat_name = f"c3s-cmip6-decadal_{version}.csv.gz"
cat_path = f"../intake/catalogs/c3s-cmip6-decadal/{cat_name}"

df_new.to_csv(cat_path, index=False, compression="gzip")
```
