# CMIP6 Decadal catalog

```python
import intake
import pandas as pd
```

```python
#cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/update-cmip6-decdal-ec-earth_2023-05-25/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

## Load combined cmip6 decadal catalog

```python
df_cmip6_decadal = cat['c3s-cmip6-decadal'].read()
df_cmip6_decadal
```

```python
df_cmip6_decadal.nunique()
```

```python
df_cmip6_decadal.institution_id.unique()
```

```python
df_dwd = df_cmip6_decadal.loc[df_cmip6_decadal.institution_id == 'EC-Earth-Consortium']
df_dwd.nunique()
```

## Load cmip6 decadal DWD catalog

```python
df_cmip6_decadal_dwd = cat['c3s-cmip6-decadal-dwd'].read()
df_cmip6_decadal_dwd
```

```python
df_cmip6_decadal_dwd.nunique()
```

## Filter ensemble members

```python
df = df_cmip6_decadal.loc[
    (df_cmip6_decadal.prefix == 'c3s-cmip6-decadal') &
    (df_cmip6_decadal.member_year == 's2016') &
    (df_cmip6_decadal.experiment_id=="dcppA-hindcast") &
    (df_cmip6_decadal.table_id=="Amon") &
    (df_cmip6_decadal.variable_id=="tas") &
    (df_cmip6_decadal.institution_id=="MPI-M")
]
l = list(df.member_id)
l.sort()
l
```
