# CMIP6 Decadal catalog: update EC_EARTH

```python
import intake
import pandas as pd
```

```python
#cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/update-cmip6-decdal-ec-earth_2023-05-25/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

## Load combined cmip6 decadal catalog

```python
df_cmip6_decadal = cat['c3s-cmip6-decadal'].read()
df_cmip6_decadal
```

```python
df_cmip6_decadal.nunique()
```

```python
df_cmip6_decadal.institution_id.unique()
```

```python
df = df_cmip6_decadal.loc[df_cmip6_decadal.institution_id == 'EC-Earth-Consortium']
df.nunique()
```

### drop old ec-earth

```python
df = df_cmip6_decadal.loc[df_cmip6_decadal.institution_id != 'EC-Earth-Consortium']
df
```

```python
df.institution_id.unique()
```

## Load EC-EARTH catalog

```python
df_cmip6_decadal_ec_earth = cat['c3s-cmip6-decadal-ec-earth'].read()
df_cmip6_decadal_ec_earth
```

## replace ec-earth datasets

```python
df = df_cmip6_decadal.loc[df_cmip6_decadal.institution_id != 'EC-Earth-Consortium']
df
```

```python
df_new = pd.concat([df, df_cmip6_decadal_ec_earth])
df_new
```

```python
df_new.institution_id.unique()
```

## Write new catalog

```python
from datetime import datetime

last_updated = datetime.now().utcnow()
version = last_updated.strftime('v%Y%m%d')
cat_name = f"c3s-cmip6-decadal_{version}.csv.gz"
cat_path = f"../intake/catalogs/c3s-cmip6-decadal/{cat_name}"

df_new.to_csv(cat_path, index=False, compression="gzip")
```
