# Reformat CMIP6 Decadal catalog

add `member_year` and `member_ensemble` columns from `member_id`

```python
import intake
import pandas as pd
```

```python
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

## Load combined cmip6 decadal catalog

```python
df_cmip6_decadal = cat['c3s-cmip6-decadal'].read()
df_cmip6_decadal
```

```python
df_cmip6_decadal.nunique()
```

```python
df_cmip6_decadal.member_id.unique()
```

```python
df = df_cmip6_decadal.loc[
    (df_cmip6_decadal.member_id == 's2016-r8i1p1f1') &
    (df_cmip6_decadal.experiment_id=="dcppA-hindcast") &
    (df_cmip6_decadal.table_id=="Amon") &
    (df_cmip6_decadal.variable_id=="tas") &
    (df_cmip6_decadal.institution_id=="MPI-M")
]
df
```

## Add new columns: member_year and member_ensemble

```python
df_cmip6_decadal['member_year'] = df_cmip6_decadal["member_id"].apply(lambda x: x.split('-')[0])
df_cmip6_decadal['member_ensemble'] = df_cmip6_decadal["member_id"].apply(lambda x: x.split('-')[1])
df_cmip6_decadal.nunique()
```

## Add new column: prefix

```python
df_cmip6_decadal['prefix'] = df_cmip6_decadal["ds_id"].apply(lambda x: x.split('.')[0])
df_cmip6_decadal.nunique()
```

## Write new Catalog

```python
from datetime import datetime

last_updated = datetime.now().utcnow()
version = last_updated.strftime('v%Y%m%d')
cat_name = f"c3s-cmip6-decadal_{version}.csv.gz"
cat_path = f"../intake/catalogs/c3s-cmip6-decadal/{cat_name}"

df.to_csv(cat_path, index=False, compression="gzip")
```
