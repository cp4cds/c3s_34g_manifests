# CMIP6 catalog: update MIROC

Remove all datasets with MIROC-ES2H.*.r1i1p1f2

On IPSL side, because the C3S-CMIP6 archive is independent from the original CMIP6 pool I ran: 

`rm -fr C3S-CMIP6/CMIP/MIROC/MIROC-ES2H/historical/r1i1p1f2`

```python
import intake
import pandas as pd
```

```python
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

## Load cmip6 catalog

```python
df_cmip6 = cat['c3s-cmip6'].read()
df_cmip6
```

```python
df_cmip6.nunique()
```

```python
df_cmip6.source_id.unique()
```

```python
df = df_cmip6.loc[df_cmip6.source_id == 'MIROC-ES2H']
df.member_id.unique()
```

### drop old MIROC-ES2H r1i1pf2

```python
# drop these datasets
df = df_cmip6.loc[(df_cmip6.source_id == 'MIROC-ES2H') & (df_cmip6.member_id == 'r1i1p1f2')]
df.nunique()
```

```python
# not ( (source_id == 'MIROC-ES2H') and (member_id == 'r1i1p1f2') )
df_new = df_cmip6.loc[(df_cmip6.source_id != 'MIROC-ES2H') | (df_cmip6.member_id != 'r1i1p1f2')]
df_new.nunique()
```

```python
# compare with original
df_cmip6.nunique() - df_new.nunique()
```

## Write new catalog

```python
from datetime import datetime

last_updated = datetime.now().utcnow()
version = last_updated.strftime('v%Y%m%d')
cat_name = f"c3s-cmip6_{version}.csv.gz"
cat_path = f"../intake/catalogs/c3s-cmip6/{cat_name}"

df_new.to_csv(cat_path, index=False, compression="gzip")
```
