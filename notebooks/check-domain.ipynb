# Intake Catalog

```python
import intake
import pandas as pd
import datetime
```

```python
# cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/update-domain/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
cats = list(cat)
cats
```

## Show all catalogs

```python
for cat_name in cats:
    df = cat[cat_name].read()
    print(f"\n{cat_name}:")
    print(df.nunique())
    # print(df.head())
    if 'url' in df:
        print(f"URL: {df.url[0]}")
```
