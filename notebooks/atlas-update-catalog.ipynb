# Update Atlas catalog

add `prefix` column to atlas catalog.

```python
import intake
import pandas as pd
```

```python
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

## Load atlas catalog

```python
df_atlas = cat['c3s-ipcc-atlas'].read()
df_atlas
```

```python
df_atlas.nunique()
```

## Add new column: prefix

```python
df_atlas['prefix'] = df_atlas["ds_id"].apply(lambda x: x.split('.')[0])
df_atlas.nunique()
```

## Show new catalog

```python
df_atlas.head()
```

## Write new Catalog

```python
from datetime import datetime

last_updated = datetime.now().utcnow()
version = last_updated.strftime('v%Y%m%d')
cat_name = f"c3s-ipcc-atlas_{version}.csv.gz"
cat_path = f"../intake/catalogs/c3s-atlas/{cat_name}"

df_atlas.to_csv(cat_path, index=False, compression="gzip")
```
